# MODELING THE DATASET

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import  RandomizedSearchCV, train_test_split

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = data.drop('target', axis=1)
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Random Forest

```python
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
y_pred2 = rfc.predict(X_test)
```

```python
acc = (accuracy_score(y_test,y_pred2))
```

```python
print('Accuracy is ',acc*100,'%')
```

```python
print("Accuracy Test score : ",round(rfc.score(X_test,y_test)*100,2),"%")
print("Accuracy Train score : ",round(rfc.score(X_train,y_train)*100,2),"%")
print("----------------------")
print("Classification Report")
print("----------------------")
clf_con = confusion_matrix(y_test,y_pred2)
print(classification_report(y_test,y_pred2))
plt.figure(figsize=(5,3))
sns.heatmap(clf_con, annot=True,cmap="YlGnBu",fmt='g')
```

# KNN

```python
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(X_train,y_train)
y_pred1 = knn.predict(X_test)
acc1 = (accuracy_score(y_test,y_pred1))
print('Accuracy is ',acc1*100,'%')
```

# Compairing Both Algorithms

```python
from matplotlib.pyplot import figure
figure(figsize=(8,5),dpi=70)
accuracy_list={'Random Forest - 98.05 %':acc,'KNN - 90.25 %':acc1}
names= list(accuracy_list.keys())
values = list(accuracy_list.values())
plt.bar(range(len(accuracy_list)) ,values,tick_label=names,)
plt.xlabel('Algorithms')
plt.ylabel('Accuracy in Percent %')
plt.show()
```
